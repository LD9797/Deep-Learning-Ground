## Funciones multivariable
Vector Gradiente: Para cada una de las siguientes funciones multivariable: grafique su superficie, calcule el vector gradiente manualmente, evaluelo y grafique el vector unitario o de largo conveniente para su visualización, en la dirección del gradiente para los dos puntos especificados (en la misma figura de la superficie). Finalmente calcule la magnitud de tal vector gradiente en cada punto.

a) (5 puntos) $f\left(x,y\right)=\sqrt{x^{2}+y^{2}}$, evaluación del gradiente en los puntos $P_{0}=\left(5.2,6.4\right)$ y $P_{1}=\left(5.2,2.3\right)$.

Gráfica de la superficie de la función:

```python
#Función multivariable original
def funcion_z(x, y):
    return np.sqrt(X**2 + Y**2)

#Calculo del Vector Gradiente (Derivadas parciales)
def dx(X,Y):
    return X / np.sqrt(X**2 + Y**2)

def dy(X,Y):
    return Y / np.sqrt(X**2 + Y**2)

def vector_gradiente(X,Y):
    return [dx(X,Y),dy(X,Y),0]


P0=[5.2,6.4]
P1=[5.2,2.3]

vector_p0=vector_gradiente(P0[0],P0[1])
vector_p1=vector_gradiente(P1[0],P1[1])

print(vector_p0)
print(vector_p1)


#Inicia el plot
N=256

x_values=np.linspace(-1,1,N)
y_values=np.linspace(-1,1,N)

X, Y = np.meshgrid(x_values,y_values)
Z = funcion_z(X,Y)

#fig=plt.figure(figsize=(7,7))
fig=plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#plt.quiver(origin, origin, gradientVector[:,0], gradientVector[:,1], color=['r','b','g'], scale=21)
origin=[0,0,0]
ax.quiver(origin[0], origin[1], origin[2], vector_p0[0],vector_p0[1],vector_p0[2], color='r')
ax.quiver(origin[0], origin[1], origin[2], vector_p1[0],vector_p1[1],vector_p1[2], color='b')

ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth=10,antialiased=False)

#Magnitud de los vectores:
print("La magnitud del vector P0", np.linalg.norm(vector_p0))
print("La magnitud del vector P1", np.linalg.norm(vector_p1))
```
